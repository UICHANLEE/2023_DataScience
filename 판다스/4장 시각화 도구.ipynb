### 그래프 꾸미기
- x축 눈금 라벨의 글씨가 서로 겹쳐 잘 보이지 않는 문제를 해결하기 위한 방법
1. figure() 함수로 그림틀의 가로 사이즈를 더 크게 설정
2. xticks() 함수를 ㅣㅇ용하여 x축 눈금을 회전

```python
#그림 사이즈 짖어 (가로 14인치, 세로 5인치)
plt.figure(figsize=(14, 5))
# x축 눈금 라벨 회전하기
plt.xticks(rotation = 45)

plt.plot(sr_one.index, sr_one.values)
plt.title('서울 -> 경기 인구 이동')

plt.xlabel('기간')
plt.ylabel('이동 인구수')
plt.show()
```

- 스타일 서식 지정, 색, 폰트 등 디자인적 요소를 사전에 지정된 스타일로 일괄 변경 가능

```python
# 그래프 스타일 목록
print(plt.style.available)
```

```python
# 스타일 서식 지정
plt.style.use('ggplot')

#그림 사이즈 지정
plt.figure(figsize = (14, 5))

# x축 눈금 라벨 회전하기
plt.xticks(size = 10, rotation = 45)

# x, y 축 데이터를 plot 함수에 입력
plt.plot(sr_one.index, sr_one.values, marker = 'X', markersize = 10) #마커 표시 추가

plt.title('서울 -> 경기 인구 이동', size = 30) # 차트 제목
plt.xlabel('기간', size = 20) # x축 이름
plt.ylabel('이동 인구수', size = 20) # y축 이름

plt.legend(labels = ['서울 -> 경기'], loc = 'best', fontsize = 15) # 범례표시

plt.show()
```

### 화면 분할하여 그래프 여러개 그리기 - axe객체 활용


```python
# 스타일 서식 지정
plt.style.use('ggplot')

#그림 사이즈 지정
fig = plt.figure(figsize = (10, 10))
ax1 = fig.add_subplot(2, 1, 1) # add_subplot(행의 갯수, 열의 갯수, n번째 위치)
ax2 = fig.add_subplot(2, 1, 2)

# axe 객체에 plot 함수로 그래프 출력
ax1.plot(sr_one, 'o', markersize = 5)
ax2.plot(sr_one, marker = '^', markerfacecolor = 'green', markersize = 5,
         color = 'olive', linewidth = 2, label = '서울 -> 경기') # linewidth : 선 두께
ax2.legend(loc = 'best') #loc는 범례의 위치

# y축 범위 지정 (최소, 최대)
ax1.set_ylim(50000, 800000)
ax2.set_ylim(50000, 800000)

# 축 눈금 라벨 지정 및 75 도 회전
ax1.set_xticklabels(sr_one.index, rotation = 75)
ax2.set_xticklabels(sr_one.index, rotation = 75)

plt.show()
```

```python
# 같은 화면에 여러개의 그래프 그리기
col_years_lst = list(range(1970, 2018))
col_years = []
for i in col_years_lst:
    col_years.append(str(i))
col_years[:10]
```

```python
df_3 = df_seoul.loc[['충청남도', '경상북도', '강원도'], col_years]
df_3
```

```python
plt.style.use('ggplot')

fig = plt.figure(figsize = (20, 5))
ax = fig.add_subplot(1, 1, 1)

ax.plot(col_years, df_3.loc['충청남도', :], marker = 'o', markerfacecolor = 'green',
       markersize = 10, color = 'olive', linewidth = 2, label = '서울 -> 충남')
ax.plot(col_years, df_3.loc['경상북도', :], marker = 'o', markerfacecolor = 'blue',
       markersize = 10, color = '#fbff8f', linewidth = 2, label = '서울 -> 경북')
ax.plot(col_years, df_3.loc['강원도', :], marker = 'o', markerfacecolor = 'red',
       markersize = 10, color = 'magenta', linewidth = 2, label = '서울 -> 강원')

ax.legend(loc = 'best')

ax.set_title('서울 -> 충남, 경북, 강원 인구 이동', size = 20)

ax.set_xlabel('기간', size = 12)
ax.set_ylabel('이동 인구수', size = 12)

ax.set_xticklabels(col_years, rotation = 90)

ax.tick_params(axis = "x", labelsize = 10)
ax.tick_params(axis = "y", labelsize = 10)

plt.show()
```

### 2 Seaborn 라이브러리 - 고급 그래프 도구
- Seaborn은 matplotlib의 기능과 스타일을 확장한 파이썬 시각화 도구

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.info()
```

### 회귀선이 있는 산점도
- regplot() 함수는 서로 다른 2개의 연속 변수 사이의 산점도를 그리고 

```python
# 회귀선이 있는 산점도
sns.set_style('darkgrid')
fig = plt.figure(figsize = (15, 5))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

sns.regplot(x = 'age',
           y = 'fare', 
           data = titanic,
            line_kws = {'color' : '#c72d1b'},
            scatter_kws = {'color' : '#FDD20E'},
           ax = ax1,
           fit_reg = True)

sns.regplot(x = 'age',
           y = 'fare', 
           data = titanic,
           scatter_kws = {'color' : '#815854'},
           ax = ax2,
           fit_reg = False)

plt.show()
```

### 히스토그램/ 커널 밀도 그래프
- 단변수 데이터의 분포를 확인할 때, distplot()함수를 이용
- 기본값으로 히스토그램과 커널밀도함수를 그래프로 출력.
- 그래프와 x축 사이의 면적이 1이 되도록 그리는 밀도 분포 함수


```python
fig = plt.figure(figsize = (15, 5))
ax1 = fig.add_subplot(1, 3, 1)
ax2 = fig.add_subplot(1, 3, 2)
ax3 = fig.add_subplot(1, 3, 3)

sns.distplot(titanic['fare'], ax = ax1)
sns.distplot(titanic['fare'], hist=False, ax = ax2)
sns.distplot(titanic['fare'], kde=False, ax = ax3)

ax1.set_title('titanic fare - hist/ked')
ax2.set_title('titanic fare - ked')
ax3.set_title('titanic fare - hist')

plt.show()
```

### 히트맵
- 2개의 범주형 변수를 매트릭스 형태로 분류
- cbar = False or cbar = True 옵션으로 컬러바를 표시할 수 있음

```python
table = titanic.pivot_table(index = ['sex'], columns = ['class'], aggfunc = 'size')

table
```

```python
sns.heatmap(table,
           annot = True, fmt = 'd',
           cmap = 'coolwarm',
           linewidth = .5,
           cbar = False)

plt.show()
```

```python
sns.set_style('whitegrid')

fig = plt.figure(figsize = (15, 5))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

sns.stripplot(x = "class",
             y = "age",
             data = titanic,
             ax = ax1)

sns.swarmplot(x = "class",
             y = "age",
             data = titanic,
             ax = ax2)

ax1.set_title('Strip Plot')
ax2.set_title('Swarm Plot')
plt.show()
```

### 빈도그래프

- countplot()메소드는 각 범주에 속하는 데이터의 개수를 막대 그래프로 나타냄

```python
sns.set_style('whitegrid')

fig = plt.figure(figsize = (15, 5))
ax1 = fig.add_subplot(1, 3, 1)
ax2 = fig.add_subplot(1, 3, 2)
ax3 = fig.add_subplot(1, 3, 3)

sns.countplot(x = 'class', palette = 'Set1', data = titanic, ax = ax1)
# hue에 'who'추가
sns.countplot(x = 'class', hue = 'who', palette = 'Set2', data = titanic, ax = ax2)
# dodge = False -> stacked와 같은 효과
sns.countplot(x = 'class', hue = 'who', palette = 'Set3', dodge = False, data = titanic, ax = ax3)

ax1.set_title('titanic class')
ax2.set_title('titanic class - who')
ax3.set_title('titanic class - who(stacked)')
```

### 박스/바이올린 그래프
- 박스 플롯은 범주형 데이터 분포와 주요 통계지표를 함께 제공
- 커널밀도 함수를 y축방향에 추가


```python
sns.set_style('whitegrid')

fig = plt.figure(figsize = (15, 10))
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)

sns.boxplot(x = 'alive', y = 'age', data = titanic, ax = ax1)
sns.boxplot(x = 'alive', y = 'age', hue = 'sex', data = titanic, ax = ax2)
sns.violinplot(x = 'alive', y = 'age', data = titanic, ax = ax3)
sns.violinplot(x = 'alive', y = 'age', hue = 'sex', data = titanic, ax = ax4)

plt.show()
```

### 조인트 그래프
- jointplot() 메소드는 산점도를 기본으로 표시하고, x-y축에 각 변수에 대한 히스토그램을 동시에 보여줌
- 두 변수의 관계와 데이터가 분산되어 있는 정도를 한눈에 파악하기 좋음
- 회귀선 추가 (kind = 'reg')
- 육각 산점도 (kind = 'hex')
- 커널 밀집 그래프 (kind = 'kde')

```python
sns.set_style('darkgrid')

j1 = sns.jointplot(x = 'fare', y = 'age', data = titanic)
j2 = sns.jointplot(x = 'fare', y = 'age', kind = 'reg', data = titanic)
j3 = sns.jointplot(x = 'fare', y = 'age', kind = 'hex',data = titanic)
j4 = sns.jointplot(x = 'fare', y = 'age', kind = 'kde',data = titanic)

j1.fig.suptitle('titanic fare - scatter', size = 15)
j2.fig.suptitle('titanic fare - reg', size = 15)
j3.fig.suptitle('titanic fare - hex', size = 15)
j4.fig.suptitle('titanic fare - kde', size = 15)

plt.show()
```

### 조건을 적용하여 화면을 그리드로 분할하기
- FacetGrid() 함수는 행, 열 방향으로 서로 다른 조건을 적용하여 서브플롯을 만듦
- 서브 플롯에 적용할 그래프 종류를 map()메소드를 사용하여 그리드에 전달

```python
sns.set_style('whitegrid')
#조건에 따라 그리드 나누기
g = sns.FacetGrid(data = titanic, col = 'who', row = 'survived')
# 그래프 적용하기
g = g.map(plt.hist, 'age')
```

```python
seoul_map = folium.Map(location = [37.55, 126.98], tiles = 'Stamen Terrain',
                       zoom_start = 12)

# 대학교 위치정보를 Marker표시
for name in df.index:
    lat = df.loc[name]['위도']
    lng = df.loc[name]['경도']
    folium.Marker([lat, lng], popup = name).add_to(seoul_map)
seoul_map.save("C:/sample_data/서울 대학교 지도1.html")
```

```python
seoul_map1 = folium.Map(location = [37.55, 126.98],
                       tiles = 'Stamen Toner',
                       zoom_start = 12)
# 대학교 리스트를 데이터 프레임 변환
df = pd.read_excel("C:/sample_data/서울지역 대학교 위치.xlsx")
df.set_index(df.columns[0], inplace = True)

for name in df.index:
    lat = df.loc[name]['위도']
    lng = df.loc[name]['경도']
    folium.CircleMarker([lat, lng], radius = 5,
                       color = 'brown', fill = True,
                       fill_color = 'coral',
                       fill_opacity = 0.7,
                       popup = name).add_to(seoul_map1)
seoul_map1.save("C:/sample_data/서울 대학교 지도2.html")
```

```python
import json

file_path = 'C:/sample_data/경기도인구데이터.xlsx'
df = pd.read_excel(file_path, index_col = '구분')
df.columns = df.columns.map(str)

# 경기도 시군구 경계 정보를 가진 geo-json파일 불러오기
geo_path = 'C:/sample_data/경기도행정구역경계.json'

try:
    geo_data = json.load(open(geo_path, encoding = 'utf-8'))
except:
    geo_data = json.load(open(geo_path, encoding = 'utf-8-sig'))

g_map = folium.Map(location = [37.5502, 126.982],
                 tiles = 'Stamen Terrain', zoom_start = 9)

year = '2017'

folium.Choropleth(geo_data = geo_data,
                 data = df[year],
                 columns = [df.index, df[year]],
                 fill_color = 'Set3', fill_opacity = 0.7, line_opacity = 0.3,
                 threshold_scale = [10000, 100000, 300000, 500000, 700000],
                  key_on = 'feature.properties.name',
                 ).add_to(g_map)

g_map.save("C:/sample_data/gyonggi_population_" + year + '.html')
g_map
```
